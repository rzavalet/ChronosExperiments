```python
import numpy as np
import scipy.stats
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

from IPython import display
from ipywidgets import interact, widgets

%matplotlib inline
```

```python
experiments = [100, 150, 200, 250, 300, 350]
base_tpm = []
base_ttpm = []
base_rate_ttpm = []
base_avg_service_delay = []

ac_tpm = []
ac_ttpm = []
ac_rate_ttpm = []
ac_avg_service_delay = []

aup_tpm = []
aup_ttpm = []
aup_rate_ttpm = []
aup_avg_service_delay = []

ac_aup_tpm = []
ac_aup_ttpm = []
ac_aup_rate_ttpm = []
ac_aup_avg_service_delay = []

for i in experiments:
    
    filename_base = '/home/rzavalet/stats/base_' + str(i) + '.csv'
    print("Experiment:", i, "filename:", filename_base)
    stats_base = pd.read_csv(filename_base)
    stats_base['ttpm_rate'] = stats_base['ttpm'] / stats_base['tpm']
    
    base_tpm.append(stats_base['tpm'].median())
    base_ttpm.append(stats_base['ttpm'].median())
    base_rate_ttpm.append(stats_base['ttpm_rate'].median())
    base_avg_service_delay.append(stats_base['average_service_delay_ms'].median())
    
    
    filename_ac = '/home/rzavalet/stats/ac_' + str(i) + '.csv'
    print("Experiment:", i, "filename:", filename_ac)
    stats_ac = pd.read_csv(filename_ac)
    stats_ac['ttpm_rate'] = stats_ac['ttpm'] / stats_ac['tpm']
    
    ac_tpm.append(stats_ac['tpm'].median())
    ac_ttpm.append(stats_ac['ttpm'].median())
    ac_rate_ttpm.append(stats_ac['ttpm_rate'].median())
    ac_avg_service_delay.append(stats_ac['average_service_delay_ms'].median())
    
    
    filename_aup = '/home/rzavalet/stats/aup_' + str(i) + '.csv'
    print("Experiment:", i, "filename:", filename_aup)
    stats_aup = pd.read_csv(filename_aup)
    stats_aup['ttpm_rate'] = stats_aup['ttpm'] / stats_aup['tpm']
    
    aup_tpm.append(stats_aup['tpm'].median())
    aup_ttpm.append(stats_aup['ttpm'].median())
    aup_rate_ttpm.append(stats_aup['ttpm_rate'].median())
    aup_avg_service_delay.append(stats_aup['average_service_delay_ms'].median())
    
    
    filename_ac_aup = '/home/rzavalet/stats/ac_aup_' + str(i) + '.csv'
    print("Experiment:", i, "filename:", filename_ac_aup)
    stats_ac_aup = pd.read_csv(filename_ac_aup)
    stats_ac_aup['ttpm_rate'] = stats_ac_aup['ttpm'] / stats_ac_aup['tpm']
    
    ac_aup_tpm.append(stats_ac_aup['tpm'].median())
    ac_aup_ttpm.append(stats_ac_aup['ttpm'].median())
    ac_aup_rate_ttpm.append(stats_ac_aup['ttpm_rate'].median())
    ac_aup_avg_service_delay.append(stats_ac_aup['average_service_delay_ms'].median())
    
df_tpm = pd.DataFrame({'threads':experiments,
                    'base':base_tpm,
                      'ac':ac_tpm,
                      'aup':aup_tpm,
                      'ac_aup':ac_aup_tpm})

df_ttpm = pd.DataFrame({'threads':experiments,
                        'base':base_ttpm,
                      'ac':ac_ttpm,
                      'aup':aup_ttpm,
                      'ac_aup':ac_aup_ttpm})

df_rate_ttpm = pd.DataFrame({'threads':experiments,
                             'base':base_rate_ttpm,
                      'ac':ac_rate_ttpm,
                      'aup':aup_rate_ttpm,
                      'ac_aup':ac_aup_rate_ttpm})

df_avg_service_delay = pd.DataFrame({'threads':experiments,
                                     'base':base_avg_service_delay,
                      'ac':ac_avg_service_delay,
                      'aup':aup_avg_service_delay,
                      'ac_aup':ac_aup_avg_service_delay})
```

```python
plot_style='o-'
line_styles=['o-','x-', '+-', 's-']
```

```python
# Plot the Number of transactions per minute for the different methods
df_tpm.plot('threads',kind='line', style=line_styles, figsize=(10, 5), grid=True,)

pp.title('Transactions per minute (TPM)')
pp.ylabel('TPM')
pp.xlabel('Number of Threads')
```

```python
# Plot the Number of timely transactions per minute for the different methods
df_ttpm.plot('threads',kind='line', style=line_styles, figsize=(10, 5), grid=True,)

pp.title('Timely Transactions per minute (TPM)')
pp.ylabel('TTPM')
pp.xlabel('Number of Threads')
```

```python
# Plot the Number of timely transactions per minute for the different methods
df_rate_ttpm.plot('threads',kind='line', style=line_styles, figsize=(10, 5), grid=True, ylim=(0,1))

pp.title('Rate of Timely Transactions per minute (TPM)')
pp.ylabel('TTPM rate')
pp.xlabel('Number of Threads')
```

```python
# Plot the Number of timely transactions per minute for the different methods
df_avg_service_delay.plot('threads',kind='line', style=line_styles, figsize=(10, 5), grid=True,)

pp.title('Average Service Delay')
pp.ylabel('milliseconds')
pp.xlabel('Number of Threads')
```
